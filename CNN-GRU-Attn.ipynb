```python

"""
CNN + GRU + Attention Model for Alzheimer's Detection.

This notebook implements a model combining a CNN feature extractor with
a Gated Recurrent Unit (GRU) and a Self-Attention layer for classification
of Alzheimer's MRI images.
"""


import tensorflow as tf
from tensorflow.keras import layers, Model, regularizers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import os
import cv2 
from sklearn.preprocessing import label_binarize
from itertools import cycle


os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
tf.get_logger().setLevel('ERROR')


DATASET_PATH = "/kaggle/input/augmentedalzheimer/AugmentedAlzheimerDataset" # Adjust path if needed
IMG_HEIGHT = 128
IMG_WIDTH = 128
BATCH_SIZE = 32
NUM_CLASSES = 4
CLASS_NAMES = ['NonDemented', 'mildDementia', 'moderateDementia', 'severeDementia']
SEED = 123


print("Loading training data...")

train_dataset = tf.keras.utils.image_dataset_from_directory(
    DATASET_PATH,
    validation_split=0.2,
    subset="training",
    seed=SEED,
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    color_mode='grayscale', 
    class_names=CLASS_NAMES
)

print("\nLoading validation data...")

validation_dataset = tf.keras.utils.image_dataset_from_directory(
    DATASET_PATH,
    validation_split=0.2,
    subset="validation",
    seed=SEED,
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    color_mode='grayscale',
    class_names=CLASS_NAMES
)

class_names_check = train_dataset.class_names
print("\nFound classes:", class_names_check)
if class_names_check != CLASS_NAMES:
    print("Warning: Class names mismatch.")



normalization_layer = layers.Rescaling(1./255)
train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y), num_parallel_calls=tf.data.AUTOTUNE)
validation_dataset = validation_dataset.map(lambda x, y: (normalization_layer(x), y), num_parallel_calls=tf.data.AUTOTUNE)


AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.cache().prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.cache().prefetch(buffer_size=AUTOTUNE)
print("\nDataset preprocessing and optimization complete.")


class SelfAttention(layers.Layer):
   
    def __init__(self, units, **kwargs):
        super(SelfAttention, self).__init__(**kwargs)
        self.units = units
        self.W1 = layers.Dense(units, activation='tanh', name='Attention_W1')
        self.W2 = layers.Dense(1, name='Attention_W2')

    def call(self, inputs):
       
        score = self.W2(self.W1(inputs))

        
        attention_weights = tf.nn.softmax(score, axis=1)

        
        context_vector = attention_weights * inputs

        
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector
    
    def get_config(self):
        config = super().get_config()
        config.update({"units": self.units})
        return config


def build_cnn_gru_attention_model(input_shape, num_classes, gru_units=128, attention_units=128):
    
    inputs = layers.Input(shape=input_shape, name='Input_Image')

    
    x = layers.Conv2D(32, (3, 3), padding='same', activation='relu', name='Conv1')(inputs)
    x = layers.MaxPooling2D((2, 2), name='Pool1')(x)
    x = layers.Conv2D(64, (3, 3), padding='same', activation='relu', name='Conv2')(x)
    x = layers.MaxPooling2D((2, 2), name='Pool2')(x)
    last_conv = layers.Conv2D(128, (3, 3), padding='same', activation='relu', name='last_conv')(x)
    
    sequence_length = last_conv.shape[1]
    feature_dim = last_conv.shape[2] * last_conv.shape[3]
    reshaped_for_rnn = layers.Reshape((sequence_length, feature_dim), name='Reshape_for_RNN')(last_conv)
    
    gru_out = layers.GRU(
        units=gru_units,
        return_sequences=True, # Must be True to feed into Attention
        kernel_regularizer=regularizers.l2(0.001),
        name='GRU_Layer'
    )(reshaped_for_rnn)
    
    attention_output = SelfAttention(units=attention_units, name='Self_Attention')(gru_out)
    
    x = layers.Dropout(0.5, name='Dropout1')(attention_output)
    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001), name='Dense1')(x)
    x = layers.Dropout(0.5, name='Dropout2')(x)
    outputs = layers.Dense(num_classes, activation='softmax', name='Output_Softmax')(x)

    model = Model(inputs=inputs, outputs=outputs, name='CNN_GRU_Attention_Model')
    return model


input_shape_cnn_gru = (IMG_HEIGHT, IMG_WIDTH, 1)
model_cnn_gru_att = build_cnn_gru_attention_model(input_shape_cnn_gru, NUM_CLASSES, gru_units=128, attention_units=128)

model_cnn_gru_att.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\n--- CNN+GRU+Attention Model Summary ---")
model_cnn_gru_att.summary(line_length=100)


EPOCHS = 30 # Increased epochs, relying on early stopping
PATIENCE = 3 # Patience for early stopping

early_stopping_gru = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy',
    patience=PATIENCE,
    restore_best_weights=True,
    verbose=1
)

print("\nStarting model training (CNN + GRU + Attention)...")
history_cnn_gru_att = model_cnn_gru_att.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=EPOCHS,
    callbacks=[early_stopping_gru]
)
print("Model training complete.")


def plot_history(history, model_name):
    
    if not history or not history.history:
        print("No training history found to plot.")
        return

    acc = history.history.get('accuracy', [])
    val_acc = history.history.get('val_accuracy', [])
    loss = history.history.get('loss', [])
    val_loss = history.history.get('val_loss', [])

    if not acc or not val_acc or not loss or not val_loss:
        print("History object missing required keys.")
        return

    epochs_range = range(len(acc))

    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy', marker='o', linestyle='-')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy', marker='o', linestyle='--')
    plt.legend(loc='lower right')
    plt.title(f'{model_name} - Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss', marker='o', linestyle='-')
    plt.plot(epochs_range, val_loss, label='Validation Loss', marker='o', linestyle='--')
    plt.legend(loc='upper right')
    plt.title(f'{model_name} - Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)

    plt.suptitle(f'Model Training History ({model_name})', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

print("\nPlotting training history...")
plot_history(history_cnn_gru_att, "CNN+GRU+Attention")


print("\nGenerating predictions on validation set...")
y_pred_probs_cnn_gru = model_cnn_gru_att.predict(validation_dataset)
y_pred_cnn_gru = np.argmax(y_pred_probs_cnn_gru, axis=1)
y_true_cnn_gru = np.concatenate([y for x, y in validation_dataset], axis=0)
print("Predictions generated.")


print("\nClassification Report (CNN+GRU+Attention):\n")
try:
    print(classification_report(y_true_cnn_gru, y_pred_cnn_gru, target_names=CLASS_NAMES, digits=4))
except ValueError as e:
    print(f"Could not generate classification report: {e}")


print("\nPlotting Confusion Matrix (CNN+GRU+Attention)...")
try:
    conf_matrix_cnn_gru = confusion_matrix(y_true_cnn_gru, y_pred_cnn_gru)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix_cnn_gru, annot=True, fmt='d', cmap='Blues',
                xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES, annot_kws={"size": 12})
    plt.title('Confusion Matrix (CNN+GRU+Attention)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
except ValueError as e:
    print(f"Could not generate confusion matrix: {e}")

print("\nCalculating and plotting ROC AUC Curves (CNN+GRU+Attention)...")
try:
   
    y_true_binarized_cnn_gru = label_binarize(y_true_cnn_gru, classes=range(NUM_CLASSES))

    if y_true_binarized_cnn_gru.shape[1] != NUM_CLASSES:
         raise ValueError(f"Binarized labels shape mismatch. Expected {NUM_CLASSES} classes.")

    fpr_cnn_gru = dict()
    tpr_cnn_gru = dict()
    roc_auc_cnn_gru = dict()

    
    for i in range(NUM_CLASSES):
        if np.sum(y_true_binarized_cnn_gru[:, i]) > 0: # Check if class has samples
             fpr_cnn_gru[i], tpr_cnn_gru[i], _ = roc_curve(y_true_binarized_cnn_gru[:, i], y_pred_probs_cnn_gru[:, i])
             roc_auc_cnn_gru[i] = auc(fpr_cnn_gru[i], tpr_cnn_gru[i])
        else:
             fpr_cnn_gru[i], tpr_cnn_gru[i], roc_auc_cnn_gru[i] = None, None, float('nan')

    
    plt.figure(figsize=(10, 8))
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green'])

    for i, color in zip(range(NUM_CLASSES), colors):
        if fpr_cnn_gru[i] is not None:
            plt.plot(fpr_cnn_gru[i], tpr_cnn_gru[i], color=color, lw=2,
                     label=f'ROC curve of class {CLASS_NAMES[i]} (area = {roc_auc_cnn_gru[i]:0.3f})')
        else:
            plt.plot([], [], color=color, lw=2, label=f'ROC curve of class {CLASS_NAMES[i]} (No samples)')

    plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier (AUC = 0.5)')
    
    valid_aucs = [roc_auc_cnn_gru[i] for i in range(NUM_CLASSES) if not np.isnan(roc_auc_cnn_gru[i])]
    if valid_aucs:
        macro_auc = np.mean(valid_aucs)
        plt.plot([], [], ' ', label=f'Macro Average AUC = {macro_auc:0.3f}')

    plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title('ROC - CNN+GRU+Attention (One-vs-Rest)')
    plt.legend(loc="lower right"); plt.grid(True); plt.show()

except Exception as e:
    print(f"An error occurred during ROC AUC calculation/plotting: {e}")


MODEL_SAVE_PATH_GRU = "dementia_cnn_gru_attention_model_final.h5"
try:
    model_cnn_gru_att.save(MODEL_SAVE_PATH_GRU)
    print(f"\nModel saved successfully as {MODEL_SAVE_PATH_GRU}")
except Exception as e:
    print(f"\nError saving model: {e}")



def get_img_array(img_path, size=(IMG_HEIGHT, IMG_WIDTH)):
    
    try:
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=size, color_mode="grayscale")
        array = tf.keras.preprocessing.image.img_to_array(img)
        array = np.expand_dims(array, axis=0)
        array = normalization_layer(array) # Use the same normalization
        return array
    except Exception as e:
        print(f"Error loading/processing image {img_path}: {e}")
        return None

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    """ Computes the Grad-CAM heatmap """
    grad_model = Model(
        model.inputs, [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    grads = tape.gradient(class_channel, last_conv_layer_output)

    if grads is None:
        print(f"Warning: Gradients are None for layer {last_conv_layer_name}.")
        return None

    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output * pooled_grads
    heatmap = tf.reduce_sum(heatmap, axis=-1)
    heatmap = tf.maximum(heatmap, 0)
    max_val = tf.reduce_max(heatmap)
    if max_val == 0:
        heatmap = heatmap.numpy()
    else:
        heatmap = (heatmap / max_val).numpy()
    return heatmap

def overlay_heatmap_on_image(heatmap, img_path, alpha=0.5, colormap=cv2.COLORMAP_JET):
    """ Overlays the heatmap on the original image """
    try:
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Could not read image {img_path} with OpenCV.")
            return None
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

        if len(heatmap.shape) > 2: heatmap = heatmap.squeeze()
        heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
        heatmap = np.maximum(heatmap, 0)
        heatmap = heatmap / (np.max(heatmap) + 1e-10)
        heatmap = np.uint8(255 * heatmap)
        
       
        heatmap[heatmap < 128] = 0 # Show only hotter areas (adjust threshold 0-255)

        heatmap_colored = cv2.applyColorMap(heatmap, colormap)
        superimposed_img = cv2.addWeighted(heatmap_colored, alpha, img, 1 - alpha, 0)
        superimposed_img_rgb = cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB)
        return superimposed_img_rgb
    except Exception as e:
        print(f"Error overlaying heatmap: {e}")
        return None


try:
    
    sample_class = 'moderateDementia' # Example class
    class_path = os.path.join(DATASET_PATH, sample_class)
    if os.path.isdir(class_path) and len(os.listdir(class_path)) > 0:
        sample_image_name = os.listdir(class_path)[0]
        image_path_gradcam = os.path.join(class_path, sample_image_name)
        print(f"Using image for Grad-CAM: {image_path_gradcam}")

        img_array_gradcam = get_img_array(image_path_gradcam)

        if img_array_gradcam is not None:
           
            
            preds_gradcam = model_cnn_gru_att.predict(img_array_gradcam)
            pred_index_gradcam = np.argmax(preds_gradcam[0])
            pred_class_name_gradcam = CLASS_NAMES[pred_index_gradcam]
            print(f"Predicted class: {pred_class_name_gradcam} (Index: {pred_index_gradcam})")

            heatmap = make_gradcam_heatmap(img_array_gradcam, model_cnn_gru_att, "last_conv", pred_index=pred_index_gradcam)

            if heatmap is not None:
                superimposed_img_gradcam = overlay_heatmap_on_image(heatmap, image_path_gradcam, alpha=0.6)

                if superimposed_img_gradcam is not None:
                    plt.figure(figsize=(10, 5))
                    plt.subplot(1, 2, 1)
                    original_img_for_plot = np.squeeze(img_array_gradcam * 255.0).astype(np.uint8)
                    plt.imshow(original_img_for_plot, cmap='gray')
                    plt.title('Original Input (Grayscale)')
                    plt.axis('off')

                    plt.subplot(1, 2, 2)
                    plt.imshow(superimposed_img_gradcam)
                    plt.title(f'Grad-CAM Overlay (Predicted: {pred_class_name_gradcam})')
                    plt.axis('off')

                    plt.tight_layout()
                    plt.show()
                else:
                    print("Could not generate superimposed image.")
            else:
                 print("Could not generate heatmap.")
        else:
             print("Could not load image array for Grad-CAM.")
            
        

    else:
        print(f"Error: Directory not found or is empty for class {sample_class} at {class_path}")

except Exception as e:
    print(f"An unexpected error occurred during Grad-CAM setup: {e}")


```
